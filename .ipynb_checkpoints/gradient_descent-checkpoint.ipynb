```python
from sklearn.datasets import make_classification
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from torch.utils.data import TensorDataset, DataLoader

# Create a toy dataset with 3 classes and 2 features
torch.manual_seed(42)
X, y = make_classification(n_samples=200, n_features=2, n_classes=3, n_redundant=0, n_clusters_per_class=1)
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)  # CrossEntropyLoss expects class indices

# DataLoader
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=10, shuffle=True)
```

```python
class MultiClassModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.w1 = nn.Parameter(torch.randn(1))  # First weight
        self.w2 = nn.Parameter(torch.randn(1))  # Second weight
        self.b1 = nn.Parameter(torch.randn(3))  # Bias for each class

    def forward(self, x):
        logits = self.w1 * x[:, 0].unsqueeze(1) + self.w2 * x[:, 1].unsqueeze(1) + self.b1  # Linear transformation
        return logits  # No softmax needed because CrossEntropyLoss applies it

model = MultiClassModel()
loss_fn = nn.CrossEntropyLoss()  # Multi-class loss function
optimizer = optim.SGD(model.parameters(), lr=0.1)

# Store values for visualization
w1_vals, w2_vals, loss_vals = [], [], []
```

```python
for epoch in range(50):  # Train for 50 steps
    for batch_x, batch_y in loader:
        optimizer.zero_grad()
        logits = model(batch_x)
        loss = loss_fn(logits, batch_y)
        loss.backward()
        optimizer.step()

        # Store weight updates
        w1_vals.append(model.w1.item())
        w2_vals.append(model.w2.item())
        loss_vals.append(loss.item())
```

```python
# Generate grid of weight (w1, w2) values
w1_range = np.linspace(-2, 2, 30)
w2_range = np.linspace(-2, 2, 30)
W1, W2 = np.meshgrid(w1_range, w2_range)
Loss = np.zeros_like(W1)

# Compute loss for each (w1, w2) pair
for i in range(W1.shape[0]):
    for j in range(W1.shape[1]):
        w1 = W1[i, j]  # Fix: Assign separately
        w2 = W2[i, j]  # Fix: Assign separately
        logits = w1 * X[:, 0].unsqueeze(1) + w2 * X[:, 1].unsqueeze(1) + model.b1  # Keep bias fixed
        loss = loss_fn(logits, y).item()
        Loss[i, j] = loss

# Convert stored values to numpy arrays
w1_vals = np.array(w1_vals)
w2_vals = np.array(w2_vals)
loss_vals = np.array(loss_vals)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot loss surface
ax.plot_surface(W1, W2, Loss, cmap='viridis', alpha=0.6)

# Plot gradient descent path
ax.plot(w1_vals, w2_vals, loss_vals, color='r', marker='o', markersize=5, label="Gradient Descent Path")

# Labels
ax.set_xlabel('Weight 1 (w1)')
ax.set_ylabel('Weight 2 (w2)')
ax.set_zlabel('Loss')
ax.set_title('3D Gradient Descent Visualization for Multi-Class Classification')
ax.legend()

plt.show()
```
